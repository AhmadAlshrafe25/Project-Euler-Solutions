```python
# A perfects square is a square of a positive number
# The first 5 squares are 1, 4, 9, 16, 25.
# The sum of the odd squares is 1 + 9 + 25 = 35
# what is the sum of odd numbers, among the first 734,000
```

```python
required_number = 734000
result = 0

for number in range(required_number):
    if number % 2 == 1:
        result = result + (number ** 2)
print(result)
```
