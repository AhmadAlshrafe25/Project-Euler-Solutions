```python
# By considering the terms in the Fibonacci sequence whose values do not exceed four million, find the sum of the even-valued terms.
current = 2
prev = 1
nex = 0
total = 0

while current < 4000000:
    if current % 2 == 0:
        total += current
    nex = current + prev
    prev = current
    current = nex
print(total)
```
